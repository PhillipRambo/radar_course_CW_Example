```python
import numpy as np
```

```python
fc = 5.5e9            # Carrier frequency [Hz]
c = 3.0e8             # Speed of light [m/s]

Vr_max_kmh = 10      # Maximum target speed [km/h]
Vr_max = Vr_max_kmh / 3.6   # Convert to m/s

B = 20e6             # Pluto SDR bandwidth [Hz]
RCS = 1.0            # Radar Cross Section [m^2]
Pt = 1e-3           # Transmit power [W]

G_dBi = 5            # Antenna gain [dBi]
NF_dB = 2.5          # Receiver noise figure [dB]
```

```python
G = 10**(G_dBi / 10)
NF = 10**(NF_dB / 10)

lam = c / fc

f_dmax = (2 * Vr_max * fc) / c

doppler_sensitivity = (2 * fc / c) / 3.6

fs_min = 2 * f_dmax
```

```python
# Constants
k = 1.38e-23         # Boltzmann constant
T0 = 290             # Standard noise temperature [K]

R = 20               # Target range [m]
T_dwell = 0.1        # Coherent integration time [s]

SNR = (Pt * T_dwell * G * G * lam**2 * RCS) / (
    (4 * np.pi)**3 * R**4 * k * T0 * NF
)

SNR_dB = 10 * np.log10(SNR)
```

```python
print("----- CW Doppler Radar Parameters -----")
print(f"Carrier frequency           : {fc/1e9:.2f} GHz")
print(f"Wavelength                  : {lam:.4f} m")
print(f"Max radial velocity         : {Vr_max:.2f} m/s ({Vr_max_kmh} km/h)")
print()

print("----- Doppler Characteristics -----")
print(f"Maximum Doppler shift       : {f_dmax:.2f} Hz")
print(f"Doppler sensitivity         : {doppler_sensitivity:.2f} Hz per km/h")
print(f"Minimum sampling frequency  : {fs_min:.2f} Hz")
print()

print("----- Radar Equation Result -----")
print(f"Target range                : {R} m")
print(f"Coherent dwell time         : {T_dwell} s")
print(f"SNR (linear)                : {SNR:.3e}")
print(f"SNR (dB)                    : {SNR_dB:.2f} dB")
```
